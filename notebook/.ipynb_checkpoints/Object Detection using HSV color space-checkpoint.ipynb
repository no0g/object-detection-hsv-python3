```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# get an image
img = cv2.imread("../img/480x360/8BIT/COLOR/img_480x360_3x8bit_RGB_color_bars_CMYKWRGB_100IRE.png")
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Display original and change to RGB
plt.subplot(121)
plt.imshow(img)
plt.title("BGR")
plt.subplot(122)
plt.imshow(img_RGB)
plt.title("RGB")
```

```python
# RGB Color Space Visualization of Original image
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors

# Set up Plot and Split R, G, B
r, g, b = cv2.split(img_RGB)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")

# Normalizing as Numpy array
pixel_colors = img.reshape((np.shape(img)[0]*np.shape(img)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()

# Scatter and Show
axis.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()

## Resource: https://realpython.com/python-opencv-color-spaces/
## We can see the value of R, G, B of each color with this visualization
```

```python
## Converting to HSV
hsv_img = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2HSV)

# Splitting H, S, V
h, s, v = cv2.split(hsv_img)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")

# Scatter and Show
axis.scatter(h.flatten(), s.flatten(), v.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Hue")
axis.set_ylabel("Saturation")
axis.set_zlabel("Value")
plt.show()

## Resource: https://realpython.com/python-opencv-color-spaces/
## We can see the value of R, G, B of each color with this visualization
```

```python
### Defining boundaries to mask original image
## Red
l_r = np.array([0,100,100])
u_r = np.array([0,255,255])
## Blue
l_b = np.array([91,100,100])
u_b = np.array([149,255,255])
## Green
l_g = np.array([31,100,100])
u_g = np.array([89,255,255])

## Mask image
mask_red = cv2.inRange(hsv_img,l_r,u_r)
mask_blue = cv2.inRange(hsv_img,l_b,u_b)
mask_green = cv2.inRange(hsv_img,l_g,u_g)

# Do bitwise AND operation to image
result_red = cv2.bitwise_and(img_RGB,img_RGB,mask=mask_red)
result_blue = cv2.bitwise_and(img_RGB,img_RGB,mask=mask_blue)
result_green = cv2.bitwise_and(img_RGB,img_RGB,mask=mask_green)
```

```python
## Show mask and result
plt.subplot(2, 2, 1)
plt.imshow(img_RGB)
plt.title('Original RGB')
plt.subplot(2, 2, 2)
plt.imshow(result_red)
plt.title('Red')
plt.subplot(2, 2, 3)
plt.imshow(result_blue)
plt.title('Blue')
plt.subplot(2, 2, 4)
plt.imshow(result_green)
plt.title('Green')
```

```python
## Real-Life Object
# get an image
img = cv2.imread("../img/keyboard1.jpg")
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Display original and change to RGB
plt.subplot(121)
plt.imshow(img)
plt.title("BGR")
plt.subplot(122)
plt.imshow(img_RGB)
plt.title("RGB")
```

```python
# RGB Color Space Visualization of Original image
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors

# Set up Plot and Split R, G, B
r, g, b = cv2.split(img_RGB)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")

# Normalizing as Numpy array
pixel_colors = img.reshape((np.shape(img)[0]*np.shape(img)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()

# Scatter and Show
axis.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()

## Resource: https://realpython.com/python-opencv-color-spaces/
## We can see the value of R, G, B of each color with this visualization
```

```python
## Converting to HSV
hsv_img = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2HSV)

# Splitting H, S, V
h, s, v = cv2.split(hsv_img)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")

# Scatter and Show
axis.scatter(h.flatten(), s.flatten(), v.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Hue")
axis.set_ylabel("Saturation")
axis.set_zlabel("Value")
plt.show()

## Resource: https://realpython.com/python-opencv-color-spaces/
## We can see the value of R, G, B of each color with this visualization
```

```python
### Defining boundaries to mask original image
## Red Keycaps 
l_r = np.array([0,100,100])
u_r = np.array([5,255,255])
## Yellow Keycaps 
l_y = np.array([15,100,100])
u_y = np.array([30,255,255])
# Blue Keycaps
l_b = np.array([100,140,100])
u_b = np.array([150,255,255])
# Green Keycaps
l_g = np.array([40,100,100])
u_g = np.array([85,255,255])
# Gray Keycaps
l_gr = np.array([0,0,0])
u_gr = np.array([179,40,255])


## Mask image
mask_red = cv2.inRange(hsv_img,l_r,u_r)
mask_yellow = cv2.inRange(hsv_img,l_y,u_y)
mask_blue = cv2.inRange(hsv_img,l_b,u_b)
mask_green = cv2.inRange(hsv_img,l_g,u_g)
mask_gray = cv2.inRange(hsv_img,l_gr,u_gr)


# Do bitwise AND operation to image
result_red = cv2.bitwise_and(img_RGB,img_RGB,mask=mask_red)
result_yellow = cv2.bitwise_and(img_RGB,img_RGB,mask=mask_yellow)
result_blue = cv2.bitwise_and(img_RGB,img_RGB,mask=mask_blue)
result_green = cv2.bitwise_and(img_RGB,img_RGB,mask=mask_green)
result_gray = cv2.bitwise_and(img_RGB,img_RGB,mask=mask_gray)
```

```python
from matplotlib.colors import hsv_to_rgb
# Show Red Boundaries
lo_square = np.full((10, 10, 3), l_r, dtype=np.uint8) / 255.0
do_square = np.full((10, 10, 3), u_r, dtype=np.uint8) / 255.0

plt.subplot(1, 2, 1)
plt.imshow(hsv_to_rgb(do_square))
plt.subplot(1, 2, 2)
plt.imshow(hsv_to_rgb(lo_square))
plt.show()

# SHow yellow boundaries
lo_square = np.full((10, 10, 3), l_y, dtype=np.uint8) / 255.0
do_square = np.full((10, 10, 3), u_y, dtype=np.uint8) / 255.0

plt.subplot(1, 2, 1)
plt.imshow(hsv_to_rgb(do_square))
plt.subplot(1, 2, 2)
plt.imshow(hsv_to_rgb(lo_square))
plt.show()

# SHow blue boundaries
lo_square = np.full((10, 10, 3), l_b, dtype=np.uint8) / 255.0
do_square = np.full((10, 10, 3), u_b, dtype=np.uint8) / 255.0

plt.subplot(1, 2, 1)
plt.imshow(hsv_to_rgb(do_square))
plt.subplot(1, 2, 2)
plt.imshow(hsv_to_rgb(lo_square))
plt.show()

# Show Green Boundaries
lo_square = np.full((10, 10, 3), l_g, dtype=np.uint8) / 255.0
do_square = np.full((10, 10, 3), u_g, dtype=np.uint8) / 255.0

plt.subplot(1, 2, 1)
plt.imshow(hsv_to_rgb(do_square))
plt.subplot(1, 2, 2)
plt.imshow(hsv_to_rgb(lo_square))
plt.show()

# Show Gray Boundaries
lo_square = np.full((10, 10, 3), l_gr, dtype=np.uint8) / 255.0
do_square = np.full((10, 10, 3), u_gr, dtype=np.uint8) / 255.0

plt.subplot(1, 2, 1)
plt.imshow(hsv_to_rgb(do_square))
plt.subplot(1, 2, 2)
plt.imshow(hsv_to_rgb(lo_square))
plt.show()
```

```python
## Show mask and result
plt.subplot(3, 2, 1)
plt.imshow(img_RGB)
plt.title('Original RGB')
plt.subplot(3, 2, 2)
plt.imshow(result_red)
plt.title('Red Keycaps')
plt.subplot(3, 2, 3)
plt.imshow(result_yellow)
plt.title('Yellow Keycaps')
plt.subplot(3, 2, 4)
plt.imshow(result_blue)
plt.title('Blue Keycaps')
plt.subplot(3, 2, 5)
plt.imshow(result_green)
plt.title('Green Keycaps')
plt.subplot(3, 2, 6)
plt.imshow(result_gray)
plt.title('Gray Keycaps')
```
